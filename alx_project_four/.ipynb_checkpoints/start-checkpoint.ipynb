```python
infrastructure_cost.head()
```

```python
infrastructure_cost["rural_adjusted"] = infrastructure_cost["unit_cost_USD"] * 1.5
infrastructure_cost.head()
```

```python
project_progress = project_progress.merge(infrastructure_cost, how="left")
project_progress.head()
```

```python
project_progress["budget"] = project_progress.apply(lambda row: row["rural_adjusted"] if "Rural" in row["town"] else row["unit_cost_USD"], axis=1)
project_progress.head()
```

```python
project_progress.head(12)
```

```python
project_progress["cost_minus_budget"] = project_progress["cost"] - project_progress["budget"]
project_progress.head()
```
