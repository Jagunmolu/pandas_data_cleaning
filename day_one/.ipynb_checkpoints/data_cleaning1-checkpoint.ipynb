```python
def clean_data(file_name):
    #extracting the excel file
    df = pd.read_excel(file_name)
    #allocating the data to each category
    df_consumer = df[df_cols[:6]]
    df_corporate = df[df_cols[:1] + df_cols[6:11]]
    df_home_office = df[df_cols[:1] + df_cols[11:]]
    #create a list to be able to loop through and clean each of the categories
    list_df_all = [df_consumer, df_corporate, df_home_office]
    #the dictionary is to rename a particular column in the df we are treating at the moment
    categories = {0: "Consumer", 1: "Corporate", 2:"Home Office"}
    #looping through the subset dataframes
    for idx, data in enumerate(list_df_all):
        #change the columns to the first rows in he data
        data.columns = data.iloc[0]
        #rename a column from "Ship Mode>>" to "Order ID"
        data.rename(columns={"Ship Mode>>": "Order ID"}, inplace=True);
        #remove the first and second rows
        data.drop([0, 1], inplace=True);
        
        data_cols = [col for col in data.columns]
        data.rename(columns={data_cols[-1]:"Total"}, inplace=True);
        data_cols = [col for col in data.columns]
        data.dropna(subset=data_cols[1:], how="all", inplace=True);
        data["Segment"] = categories[idx]
        data["Ship Mode"] = ""
        for col in data_cols[1:5]:
            data["Ship Mode"] = np.where(~data[col].isnull(), col, data["Ship Mode"])
        data.rename(columns={"Total": "Sales"}, inplace=True);
#         data = data[["Segment", "Ship Mode", "Order ID", "Sales"]]
    final_df = pd.concat(list_df_all)
    final_df.reset_index(drop=True, inplace=True);
    final_df = final_df[["Segment", "Ship Mode", "Order ID", "Sales"]]
    return final_df
```

```python
cleaned = clean_data("1.-Badly-Structured-Sales-Data-1.xlsx")
cleaned.head()
```
